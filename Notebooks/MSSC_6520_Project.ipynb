# Load Modules

```python
import numpy as np
import pandas as pd
import xgboost
from sklearn.ensemble import RandomForestRegressor
import shap
```

# Load CSV files into DataFrame

```python
# Load CSV into a pandas DataFrame
X = pd.read_csv('X.csv',header=0)
y = pd.read_csv('y.csv', header=0)
```

# Fit XGBoost Regressor

```python
model = xgboost.XGBRegressor().fit(X, y)
```

# Generate SHAP Explainer Object with the Fitted Model

```python
explainer = shap.Explainer(model)
```

# Get the SHAP Values

```python
shap_values = explainer(X)
```

# Bar Chart

```python
shap.plots.bar(shap_values)
```

# BeeSwarm Plot

```python
shap.plots.beeswarm(shap_values,)
```

# Waterfall Plot

```python
shap.plots.waterfall(shap_values[0])
```

# Heatmap Plot

```python
shap.plots.heatmap(shap_values)
```

# Fit a Random Forest Model

```python
model = RandomForestRegressor(
    n_estimators = 100,  # Increase the number of trees
    max_depth = 10,  # Limit the depth of the trees
    min_samples_split = 5,  # At least 5 samples required to split a node
    min_samples_leaf = 3,  # At least 3 samples required at a leaf node
    random_state = 42,  # Ensures reproducibility
    bootstrap = True  # Enable bootstrap sampling
)

model.fit(X, y)
```

# Generate SHAP Explainer Object with the Fitted Model

```python
explainer = shap.Explainer(model)
```

# Get the SHAP Values

```python
shap_values = explainer(X)
```

# Bar Chart

```python
shap.plots.bar(shap_values)
```

# BeeSwarm Plot

```python
shap.plots.beeswarm(shap_values,)
```

# WaterFall Plot

```python
shap.plots.waterfall(shap_values[0])
```

# HeatMap Plot

```python
shap.plots.heatmap(shap_values)
```
